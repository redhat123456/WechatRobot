```python
def generate_daily_message(str):
    tab = '{ctrl}{ENTER}'
    kaitou = '亲爱的家长：'
    ignore1 = '大家好！为了帮助孩子们更好地掌握课程内容，提升学习效率，我们决定每天为孩子们布置一些有有关英语口语的小作业。请各位家长和孩子一起认真对待，按时完成，相信在不断的练习中，孩子们会取得更大的进步。'
    ignore2 = '今天的作业安排如下：'
    task1 = '1、请孩子们仔细观看发到微信群中的视频。请引导他们认真思考并体会其中的情感和寓意。'
    task2 = '2、让孩子针对视频中的内容跟着进行配音，并录制相关视频。'
    task3 = '3、录制完成后，请大家将视频文件发到微信群里。'
    content = f"{kaitou}{tab}{ignore1}{tab}{ignore2}{tab}{task1}{tab}{task2}{tab}{task3}"
    return content
def end():
    end = '请在23点之前交到微信群里。[玫瑰][玫瑰]'
    return end
```
