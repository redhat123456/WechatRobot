```python
# pip install PyOfficeRobot
import PyOfficeRobot
import pandas as pd
```

```python
# 不同年级不同章节内容
str = '第十七章'
```

```python
# 读取 CSV 文件
df = pd.read_csv('user.csv')

# 提取用户列
users = df['users'].tolist()
users_name = df['users_name'].tolist()
users_nianji = df['users_nianji'].tolist()
# 输出提取出的用户列
df
```

```python
def generate_daily_message(str):
    tab = '{ctrl}{ENTER}'
    ignore1 = f"家长好！请提醒孩子进行今天的{str}数学作业。"
    ignore2 = '完成作业后，请将作业发到我们的群里。我会在群里检查大家的完成情况，请务必确保孩子按时并认真地完成这项任务。感谢大家的配合与支持！[玫瑰][玫瑰]'
    content = f"{ignore1}{tab}{ignore2}"
    return content
def end():
    end = '请在24点之前交到微信群里。[玫瑰][玫瑰]'
    return end
```

```python
for index, user in enumerate(users):
    # 在这里你可以根据索引index来访问数组的不同元素
    # 构造消息，包括用户名称和每日消息
    print(f"Sending message to {users[index]}")
    nianji = users_nianji[index]  # 获取用户的年级
    message = f"{users_name[index]}{generate_daily_message(str)}"
    # 发送消息
    print(message)
    PyOfficeRobot.chat.send_message(who=users[index], message=message)
    # 发送文件

    if nianji == 3:
        file_path = r'C:\Users\Tanger\Desktop\工作\code2\3.png'
    elif nianji == 4:
        file_path = r'C:\Users\Tanger\Desktop\工作\code2\4.png'
    elif nianji == 5:
        file_path = r'C:\Users\Tanger\Desktop\工作\code2\5.png'
    elif nianji == 6:
        file_path = r'C:\Users\Tanger\Desktop\工作\code2\6.png'

    print(file_path)
    PyOfficeRobot.file.send_file(who=users[index], file=file_path)
    
    #发送结尾消息
    PyOfficeRobot.chat.send_message(who=users[index], message=end())
```
