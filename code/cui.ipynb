```python
# 读取 CSV 文件
df = pd.read_csv('user2.csv')

# 提取用户列
users = df['users'].tolist()
users_name = df['users_name'].tolist()
# 输出提取出的用户列
df
```

```python
for index, user in enumerate(users):
    # 在这里你可以根据索引index来访问数组的不同元素
    # 发送催交作业的信息
    print(f"Sending message to {users[index]}")
    message = f"没有完成作业的同学请抓紧时间完成，已经完成或住校的同学忽略本条信息。[玫瑰][玫瑰]"
    PyOfficeRobot.chat.send_message(who=users[index], message=message)
```
